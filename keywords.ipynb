```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
filepath = "Data/keywords.csv"
```

```python
#create a dataframe from csv
df = pd.read_csv(filepath, delimiter=',')

list_of_rows = [list(row) for row in df.values]
type(list_of_rows)
```

```python
print(len(list_of_rows))
```

```python
print(list_of_rows[0][1])
```

```python
jd = eval(list_of_rows[1][1])
print(jd)
```

```python
#Run a loop for all rows in the spreadsheet #variables = movie_id and keyword
list_of_keywords_dataframes= []
for row in list_of_rows:
    name_list = eval(row[1])
    record_df = pd.DataFrame(name_list)
    record_df['movie_id'] = row[0]
    list_of_keywords_dataframes.append(record_df)

name_table = pd.concat(list_of_keywords_dataframes, sort=False)
name_table.reset_index(inplace=True, drop=True)
name_table.head(30)
```

```python
name_table = name_table.rename(columns={'index':'rec_key', 'id':'keyword_id', 'movie_id':'movie_id', 'name': 'keyword'})
name_table.head(30)
name_table.reset_index()
```
